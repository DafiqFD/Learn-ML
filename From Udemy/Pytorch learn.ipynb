Reshaping, stacking, squeezing and unsqueezing tensor

1. Reshape - reshape input tensore to a defined shape
2. View - return a view of an input tensor of certain shape but keep the same memory as original
3. Stacking - combine multiple tensors on top of each other or side by side
4. squeeze - remove all '1' dimension from a tensor
5. unsqueeze - add '1' dimension to a target tensor
6. permute - return a view of the input with dimension swapped in a certain way

Reproducecibility (trying to take random out of random)

How neural network learn:
random num -> tensor operation -> update random number to make better data -> again . . .

To reduce randomness in NN comes concept of **random seed**

**random seed** = flavour of randomness

```python
random_A = torch.rand(3,4)
random_B = torch.rand(3,4)

print(random_A)
print(random_B)
print(random_A == random_B)
```

```python
# reproducible
random_seed = 42
# the number is for the “starting point” of the algorithm to be
# it will move from there and generate random numbers.
torch.manual_seed(random_seed)
random_seed_c = torch.rand(3,4)
torch.manual_seed(random_seed)
random_seed_d = torch.rand(3,4)

print(random_seed_c)
print(random_seed_d)
print(random_seed_c == random_seed_d)
# 'Flavoured' Randomness
```
